```python
import pandas as pd
# Baca file CSV menjadi DataFrame
df = pd.read_csv(r"C:\Users\siaha\Documents\TETRIS 4\ProjectCapstoneDesign\DatasetProjectCapstone\PPK_Garam.csv")

# Mencari baris dengan kata kunci "Jakarta" atau "Seribu" pada kolom "Kabupaten/Kota"
keywords = ["Jakarta", "Seribu"]

# Filter DataFrame dengan baris yang mengandung kata kunci pada kolom "Kabupaten/Kota"
filtered_df = df[df['Kabupaten/Kota'].str.contains('|'.join(keywords), case=False)]

# Tampilkan hasil pencarian
print("Hasil Pencarian:")
print(filtered_df)
```

```python
# Menghitung total per tiap kolom tahun
totals_per_year = filtered_df.iloc[:, 1:].sum().tolist()

# Menampilkan hasil
print("Total per Tahun:", totals_per_year)
year =[
    '2018',
    '2019',
    '2020',
    '2021',
    '2022',
    '2023'
]
import altair
import pandas as pd

new_data = {
    'Tahun' : year,
    'Total Pengeluaran/Kapita' : totals_per_year
}
new_df = pd.DataFrame(new_data)
new_df
```

```python
import matplotlib.pyplot as plt
# Membuat figure dan axes untuk dua kolom chart
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Diagram pie
axs[0].pie(totals_per_year, labels=year, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Diagram Pie: Total Pengeluaran/Kapita per Tahun')

# Diagram garis
axs[1].plot(year, totals_per_year, marker='o', linestyle='-')
axs[1].set_title('Diagram Garis: Total Pengeluaran/Kapita per Tahun')
axs[1].set_xlabel('Tahun')
axs[1].set_ylabel('Total Pengeluaran/Kapita')

# Menampilkan chart
plt.tight_layout()
plt.show()
```
